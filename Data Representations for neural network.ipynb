- `Shape` - This is a tuple of integers that describes how many dimensions the tensors has along each axis.For example let's say a matrix has hape (3,5) and the 3D tensor has shape (3,3,5). A vector has a shape with a single element, such as (5,), whereas a scalar has a an empty shape,().

#### 8. Real-world examples of data tensors
---

The main data which we are going to encounter fall into these category:  

- Vector data: 2D tensors **(samples,features)**
- TimeSeries or sequence data: 3D tesnors of shape **(samples,timestamps,features)**
- Images: 4D tensors of shape **(samples,height,width,channels) or (samples,channels,height,width)**
- Video: 5D tensors of shape **(samples,frames,height,width,channels) or (samples,frames,channels,height,width)**  


**Now let's talk about Vector,Timeseries.Images,Video data one by one**

##### 8.1 Vector Data
---

This is the **most common dataset**, each single data point can be encoded as a vector, and thus a batch will be encoded as 2D tensor(that is an array of vectors), where the first axis is the sample axis and the second axis as the feature axis.  
Examples:  

- An actuarial dataset of people, where we consider each person's age, ZIP code, and income. Each person can be characterized as a vector of 3 values and thus the entire dataset of 100,000 people can be stored in a 2D tensor of shape (1000,3).

- A dataset of text documents, where we represnet each documents by the counts of how many times each word appear in it (out of a dictionary of 20,000 common word).Each doucment can be encoded as a vector of 20 000 values, and thus the entire dataset of 500 documents can be stored in a tensor of shape (500,20000)  

###### 8.2 TimeSeries data or sequence data  
---

Whenever time matters in your data( or the notion of sequence order), it makes sense to store it in a 3D tensors, with an explicit time axis.Each samples can be encoded as a sequence of vectors( a 2D tesnor), and thus a batch of data will be encoded as a 3D tensors.

**<u>The time axis will always be the second axis (index 1), by convention.</u>** 

Examples:
- A dataset of stock prices. Every minute, we store the **current price of the stock**, **the highest price in the past time**, and **the lowest price in the past minute**.Thus every minute is encoded as a 3D vector, an entire day of trading is encodes as 2D tesnor of shape (390,3)(there are 390 minutes in a trading day), and 250 day's worth of data can be stored in a 3D tensor of shape (250, 390,3).  



- A dataset of tweets, where we encode each tweets as a sequence of 280 characters out of an aplhabet of 128 unique characters.In this setting, each character can be encoded as a binary vector of size 128.Then each tweet can be encoded as 2D tensor of shape (280,128), and a dataset of 1 million tweets can be stored in a tensor of shape (1000000,280,128)  

##### 8.3 Image Data
---

Images typically have three dimesnions: **height**, **width** and **colour depth**.Although grayscale images have only a single color channel and could be stored in a 2D tensors, by convention image tensor are always 3D, with a one- dimesnional color channels for grayscale images. A **batch** of 128 **grayscale image** of size 256 X 256 could be stored in a tensor shape **(128,256,256,1)** and  batch of 128 **color images** can be stored in a tensor of shape **(128,256,256,3)**.

There are two convetions for shapes of images tensors: the *channels-last* convnentions and the *channels-first* conventions. Tensorflow convetions (samples,height,width,color_depth).Thenao (samples,color_depth,height,width).Keras support both formats.  

##### 8.4 Video data
---

Video data is one of the few types of real-world data for which you'll need 5D tensors.A video can be understood as a sequence of frames, each frames being a color images.Because each frame can be stored in a 3D tensor(height,width,color_depth), a sequence of frames can be stored in a 4D tensor(frames,height,width,color_depth), and thus batch of different videos can be stored in a 5D tesnor of shape (samples,frames,height,width,color_depth). 

For examples:
- A 60 second, 144 x 256 youtube video clip sampled at 4 frames per second would have 240 frames.A batch of four such video clip would be storws in a tensor of shape (4,240,144,256,3).
